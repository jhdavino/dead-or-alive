# the algorithm

```python
class logistic_reg:
    
    def __init__(self, x, y, threshold=0.5):
        self.threshold = threshold
        self.x = x
        self.y = y
        
    def fit(self): #using normal equation... i dont know if its work... 
        self.theta = np.dot(pinv(np.matmul(np.transpose(self.x), self.x)), np.matmul(np.transpose(self.x), self.y)) 
        return self.theta
        
    def predict(self, x):
        h = [1/(1+v) for v in np.exp(-np.matmul(x,self.theta))]
        self.hypotesis = []
        for v in h:
            if(v > self.threshold):
                self.hypotesis.append(1)
            else:
                self.hypotesis.append(0)
        return self.hypotesis
        #return h
        
    def error(self):
        h = [1/(1+v) for v in np.exp(-np.matmul(x,self.theta))]
        return (1/len(self.y)) * sum(np.subtract(np.matmul(y,np.log(h)), np.matmul(np.subtract(1,y), np.log(np.subtract(1, h)))))
```

```python
reg = logistic_reg(x,y,0.56)
reg.fit()
#reg.error()
```
